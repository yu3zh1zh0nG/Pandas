##### 在绘图结构中,figure创建窗口, subplot 创建子图. 所有的绘画只能在子图上进行.   plt表示当前年子图,若没有就创建一个子图. 

### 配置参数

```
axes : 设置坐标轴边界和表面颜色, 坐标刻度值大小和网格的显示
figure : 控制dpi, 边界颜色, 图形大小 和 子区(subplot)的设置
font : 字体集(font family) , 字体大小和样式设置
grid : 设置网格颜色和线性
legend : 设置图例和其中的文本的显示
line : 设置线条(颜色,线性,宽度等) 和标记
patch : 是填充2D空间的图形对象, 如多边形和圆. 控制线宽,颜色和抗锯齿设置等
savefig : 可以对保存的图形进行单独设置.
verbose : 设置matplotlib在执行期间信息输出,如silent helpful debug 
xticks和yticks : 为x,y轴的主刻度和次刻度设置颜色,大小,方向 以及标签大小
```

# 绘图步骤

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
````

````python
# 产生数据
x=np.arange(-5,5,0.1)
y=x**3
````

````python
# 创建窗口 子图
# 方法一: 只绘制添加的子图
fig=plt.figure(num=1,figsize=(15,8),dpi=80) # 开启一个窗口,同事设置大小,分辨率
ax1=fig.add_subplot(2,3,1)  # 参数: 行数 列数, 第几个
ax2=fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,4)
print(fig)
````

````python
# 方法二:全部绘制
fig,axarr=plt.subplots(2,3,figsize=(15,8),dpi=50)
ax1=axarr[0,1]
print(ax1)
````

````python
ax1.set_title('this is title') # 设置图体:  标题 plt.title
ax1.set_xlabel('x-zhou') # 设置x轴名称 
ax1.set_ylabel('y-zhou')

fig
````

````python
# ax1.axis([-6,6,-10,10]) # 设置横纵坐标范围,这个在子图中被分解为下面 两个函数
ax1.set_xlim(0,6) # 设置横轴坐标范围
ax1.set_ylim(0,9) #  设置纵轴坐标范围
# 方法一 : from matplotlib.ticker import MultipleLocator
ax1.xaxis.set_major_locator(MultipleLocator(2)) # x轴间隔2个刻度显示一个标签文本
ax1.yaxis.set_major_locator(MultipleLocator(3)) # y轴间隔3个刻度显示一个标签文本
fig
````

````python
#  方法二:
ax1.set_xticks([]) # 去除横坐标轴刻度
ax1.set_xticks([0,1,2,3,5,10]) # 自定义坐标刻度
# 设置刻度的显示文本, rotation旋转角度, fontsize字体大小
ax1.set_xticklabels(labels=[0,'a','b',3,'c',10],rotation=-30,fontsize=12)
fig
````

````python
plot1=ax1.plot(x,y,marker='o',color='g',label='legend1') #点图: marker图标
plot2=ax1.plot(x,y,linestyle='--',alpha=0.5,color='r',label='legend2') # 线图: linestyle线性,alpha透明度
ax1.legend(loc='upper right')
ax1.text(2.8,7,'y=x**3',fontsize='14')
fig
````

````python
# 添加标注   参数 : 注释文本, 指向点, 文本位置, 箭头属性  
ax1.annotate('important point',xy=(2,8),xytext=(3,1.5),arrowprops=dict(facecolor='red',shrink=0.5))
fig
````

````python
#当前窗口 添加子图 
# rect = [左,下,宽,高] 使用绝对定位
axes1 = plt.axes([.2, .3, .1,.1], facecolor='y')
fig
````

## 一个窗口多个图

````python
#一个窗口，多个图，多条数据
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-5,5,0.1)
y=x**3
sub1=plt.subplot(211,facecolor=(0.1843,0.3098,0.3098))  #将窗口分成2行1列，在第1个作图，并设置背景色
sub2=plt.subplot(212)   #将窗口分成2行1列，在第2个作图
sub1.plot(x,y)          #绘制子图
sub2.plot(x,y)          #绘制子图

axes1 = plt.axes([.2, .3, .1, .1], facecolor='y')  #添加一个子坐标系，rect=[左, 下, 宽, 高]
plt.plot(x,y)           #绘制子坐标系，
axes2 = plt.axes([0.7, .2, .1, .1], facecolor='y')  #添加一个子坐标系，rect=[左, 下, 宽, 高]
plt.plot(x,y)
plt.show()
````

````python
fig=plt.figure(1)
ax1=fig.add_subplot(1,1,1)

x=np.random.random(100)
y=np.random.random(100)
ax1.scatter(x,y,s=10,c='y',marker='o',alpha=0.5,lw=2,facecolors='none',label='legend1')  #x横坐标，y纵坐标，s图像大小，c颜色，marker图片，lw图像边框宽度
x=np.random.random(100)
y=np.random.random(100)
ax1.scatter(x,y,s=10,c='r',marker='o',alpha=0.5,lw=2,facecolors='none',label='legend2') 
ax1.legend(loc='upper right')
````
